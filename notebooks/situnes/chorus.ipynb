# Finding chorus of extracted wav files

```python
from pychorus import find_and_output_chorus
import os
import shutil
```

```python
directory = "../../data/wav"
out_directory = "../../data/wav_clipped"

clipped_30 = 0
clipped_15 = 0
unclipped = 0

for filename in os.listdir(directory):
    if (filename.startswith('1') or filename.startswith('2')):
        path = os.path.join(directory, filename)
        outpath = os.path.join(out_directory, filename)
        try:
            time = find_and_output_chorus(path, outpath, 30)
            if time == None:
                time = find_and_output_chorus(path, outpath, 15)
                if time == None:
                    shutil.copy(path, outpath)
                    unclipped += 1
                else:
                    clipped_15 += 1
            else:
                clipped_30 += 1
        except MemoryError:
            shutil.copy(path, outpath)
            unclipped += 1

print(clipped_30, clipped_15, unclipped)
        
```

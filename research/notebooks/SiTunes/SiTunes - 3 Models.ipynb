# Models

```
import pandas as pd
import numpy as np
```

```
df_stage2: pd.DataFrame = pd.read_csv("/work/itunes_processed/target_difference_2.csv")
df_stage3: pd.DataFrame = pd.read_csv("/work/itunes_processed/target_difference_3.csv")
# df: pd.DataFrame = pd.concat([df_stage2, df_stage3], axis=0)
df: pd.DataFrame = df_stage2
df = df.sort_values(by=["item_id"])
display(df)
```

```
df = df.groupby("item_id").mean()
display(df)
```

```
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()


scale_columns = [
    # Music high level metadata
    'popularity', 'loudness', 'danceability', 'energy',
    'speechiness', 'acousticness', 'instrumentalness', 'valence', 'tempo', 'duration_x',

    # Music low level metadata
    'F0final_sma_amean','F0final_sma_stddev', 
    'audspec_lengthL1norm_sma_stddev',
    'pcm_RMSenergy_sma_stddev', 
    'pcm_fftMag_psySharpness_sma_amean', 'pcm_fftMag_psySharpness_sma_stddev', 
    'pcm_zcr_sma_amean', 'pcm_zcr_sma_stddev', 

    # Situational
    'pressure', 'temperature', 'humidity','gps_speed', 

    # # Categorical discrete
    # 'key', 'general_genre_id', 'time', 'weather type' 
]


df[scale_columns] = scaler.fit_transform(df[scale_columns])
df
```

```
targets = ["arousal_change", "valence_change", "emo_pre_arousal", "emo_post_arousal", "emo_pre_valence", "emo_post_valence"]

df_target = df[targets]
df_features = df.drop(columns=targets)
```

```
display(df_features)
```

```
columns=['popularity','danceability','energy','key','valence','tempo', 'general_genre_id']
df_features = df_features[columns]
df_features
```

```
df_features = pd.get_dummies(df_features, columns=["key"])
df_features = pd.get_dummies(df_features, columns=["general_genre_id"])
df_features
```

```
display(df_target)
```

## Predicting Arousal Change

```
X = df_features
y_arousal = df_target['arousal_change']
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_arousal_train, y_arousal_test = train_test_split(X, y_arousal, test_size=0.2, random_state=42)
```

### Multiple Linear Regression

```
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_arousal_train)
```

```
def display_metrics(model, features, target):
    pred = model.predict(features)

    # The root mean squared error
    mse = mean_squared_error(target, pred)
    rmse = np.sqrt(mse)
    print("Root mean squared error: %.3f" % rmse)

    # The coefficient of determination: 1 is perfect prediction
    r2_value = r2_score(target, pred)
    print("Coefficient of determination (R-squared value): %.3f" % r2_value)
```

```
display_metrics(model=regr, features=X_train, target=y_arousal_train)
```

```
display_metrics(model=regr, features=X_test, target=y_arousal_test)
```

### Support Vector

```
from sklearn.svm import SVR
```

```
X_train, X_test, y_arousal_train, y_arousal_test = train_test_split(X, y_arousal, test_size=0.2, random_state=42)
```

```
svr = SVR(kernel='rbf', gamma=0.1)
svr.fit(X_train, y_arousal_train)
```

```
display_metrics(model=regr, features=X_train, target=y_arousal_train)
```

```
from sklearn.model_selection import GridSearchCV 

param_grid = {'C': [0.1, 1, 10, 100, 1000], 
			'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
			'kernel': ['rbf']} 

grid = GridSearchCV(SVR(), param_grid, refit = True, verbose = 3) 

grid.fit(X_train, y_arousal_train) 
```

```
print(grid.best_params_)
```

```
display_metrics(model=grid, features=X_train, target=y_arousal_train)
```

```
display_metrics(model=grid, features=X_test, target=y_arousal_test)
```

### Random Forest Regression Model

```
from sklearn.ensemble import RandomForestRegressor
```

```
X_train, X_test, y_arousal_train, y_arousal_test = train_test_split(X, y_arousal, test_size=0.2, random_state=42)
```

```
fr = RandomForestRegressor(max_depth=5, random_state=42)
fr.fit(X_train, y_arousal_train)
```

```
display_metrics(model=fr, features=X_train, target=y_arousal_train)
```

```
display_metrics(model=fr, features=X_test, target=y_arousal_test)
```

Models perform very poorly on predicting change

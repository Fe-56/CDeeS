### Preprocessing Env.json

Read env.json file into Dataframe

```
# Reading environmental conditions
def env_json_to_df(json_path: str) -> pd.DataFrame:
    env_df = pd.read_json(json_path)
    env_df = env_df.transpose()
    env_df['inter_id'] = env_df.index

    # Split weather column
    weather_cols = ['weather type', 'pressure', 'temperature', 'humidity']
    env_df[weather_cols] = env_df.apply(lambda x: x.weather, axis=1, result_type='expand')

    # Split GPS column
    gps_cols = ["gps_lat", "gps_lng", "gps_speed"]
    env_df[gps_cols] = env_df.apply(lambda x: x.GPS, axis=1, result_type='expand')

    # Drop parsed columns
    env_columns_to_drop = ["weather", "GPS"]
    env_df2 = env_df.drop(columns=env_columns_to_drop)

    # Reorder columns for cleanliness
    env_df2 = env_df2[['inter_id', 'time', 'weather type', 'pressure', 'temperature', 'humidity', "gps_lat", "gps_lng", "gps_speed"]]

    return env_df2
```

```
env_df2 = env_json_to_df("/work/SiTunes_dataset/SiTunes/Stage2/env.json")
display(env_df2)
```

```
env_df3 = env_json_to_df("/work/SiTunes_dataset/SiTunes/Stage3/env.json")
display(env_df3)
```

```
# Adding in environmental details to the dataframe
merge_env2 = pd.merge(merged_2, env_df2, left_on='inter_id', right_on='inter_id', how='inner')
display(merge_env2)
```

### Preprocessing for recommendation data

Different set of columns are considered for the recommendation task

```
def append_va_delta(df):
    # Calculate difference in arousal and valence
    df['arousal_change'] = df.apply(lambda row: row['emo_post_arousal'] - row['emo_pre_arousal'], axis=1)
    df['valence_change'] = df.apply(lambda row: row['emo_post_valence'] - row['emo_pre_valence'], axis=1)
    return df
```

Stage 2

```
columns_to_drop_rec_stage2 = [
    'general_genre', 
    'music', 
    'inter_id', 
    'item_id', 
    'timestamp', 
    'duration_y', 
    'gps_lat', 
    'gps_lng'
    ]

# Keep user_id, rating, singer, rec_type (for EDA evaluation)
```

```
df2_rec = merge_env2.drop(columns=columns_to_drop_rec_stage2)
df2_rec = df2_rec.dropna() # Missing rec_type
append_va_delta(df2_rec)
display(df2_rec)
```
